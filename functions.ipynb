```python
# Function to explore data and check

#make copy
def make_copy(dataframe):
    copy = dataframe.copy()
    if copy.empty:
        return "Your dataframe is empty!"
    else:
        return copy

#check the size of the df
def df_shape(dataframe):
    shape = dataframe.shape
    if shape != (0,0):
        return f"The shape/size of your dataframe is {shape}"
    else:
        return "Your dataframe is empty and is of size {shape}"

#get information - columns, datatypes, etc
def get_info(dataframe):
    info = dataframe.info()
    if dataframe.empty:
        return "Your dataframe is empty. Check your dataframe variable"
    else:
        return info

#Get a sample of the data
def data_sample(dataframe):
    sample = dataframe.head()
    if sample.empty:
        return "Your dataframe is empty. Check your dataframe variable"
    else:
        return sample
    
#Check for nulls and dups
def check_dups(dataframe):
    num_dups = dataframe.duplicated().sum()
    if dataframe.empty:
        return "Your dataframe is empty. Check your dataframe variable"
    else:
        return "There are {num_dups} duplicates in your dataframe"
    
#Check nulls
def check_nulls(dataframe):
    num_nulls = dataframe.isnull().sum()
    if dataframe.empty:
        return "Your dataframe is empty. Check your dataframe variable"
    else:
        return "The number of nulls per columns: ", num_nulls
    
#get summary statistics
def summary_stats(dataframe):
    stats = dataframe.describe()
    if dataframe.empty:
        return "Your dataframe is empty. Check your dataframe variable"
    else:
        return "Here are the summary statistics for your dataframe: ", stats

```
